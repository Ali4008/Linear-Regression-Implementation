```python
import numpy as np
import random
import matplotlib.pyplot as plt
random.seed(42)  # Setting seed for reproducibility
# Initialize empty lists
x_data = []
y_data = []

# Generate x_data using a for-loop
for _ in range(20):
    x_value = random.uniform(0, 20)  # Generate a random number between 0 and 20
    x_data.append(x_value)           # Append it to the x_data list

# Generate y_data using a for-loop
for x in x_data:
    y_value = 3 * x + 2 + random.uniform(-2, 2)  # Calculate y using the formula and add some noise
    y_data.append(y_value)                       # Append it to the y_data list
```

```python
# # Calculating coefficients b (slope) and a (y-intercept)
# x_mean = np.mean(x_train)
# y_mean = np.mean(y_train)
# b = np.sum((x_train - x_mean) * (y_train - y_mean)) / np.sum((x_train - x_mean)**2)
# a = y_mean - b * x_mean

# Convert to NumPy arrays
x_data = np.array(x_data)
y_data = np.array(y_data)
# Calculate mean of x and y
mean_x = np.mean(x_data)
mean_y = np.mean(y_data)

# Calculate parameters "a" and "b" (c in the equation y = ax + b)
numerator = np.sum((x_data - mean_x) * (y_data - mean_y))
denominator = np.sum((x_data - mean_x)**2)
m = numerator / denominator
c = mean_y - m * mean_x

print (m)
print (c)
```

```python
# Visualize the data points and regression line
plt.scatter(x_data, y_data, label='Data Points')
y_predict = c + m * x_data
plt.plot(x_data, y_predict, color='red', label='Regression Line')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Linear Regression')
plt.legend()
plt.show()
```

```python
# Calculate Mean Squared Error (MSE) for training data
mse_train = np.mean((y_data - y_predict)**2)

print (mse_train)
```

```python
# Separate test data
x_test=[]
y_test=[]
for _ in range(5):
  val=random.uniform(0,10)
  x_test.append(val)
for x in x_test:
  val=3*x+2+random.uniform(-2,2)
  y_test.append(val)
# Predict using the regression line
y_pred_test = c + m * np.array(x_test)

print (y_pred_test)
```

```python
# Calculate MSE for testing data
mse_test = np.mean((np.array(y_test) - y_pred_test)**2)

print (mse_test)
```

```python
# Linear Regression using Gradient Descent
theta0, theta1 = random.random(), random.random()
learning_rate = 0.001
epochs = 1000

for _ in range(epochs):
    predictions = theta0 + theta1 * x_data
    error = predictions - y_data
    theta0 -= learning_rate * np.mean(error)
    theta1 -= learning_rate * np.mean(error * x_data)

# Visualize the data points and regression line from Gradient Descent
plt.scatter(x_data, y_data, label='Data Points')
plt.plot(x_data, theta0 + theta1 * x_data, color='red', label='Regression Line (Gradient Descent)')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Linear Regression with Gradient Descent')
plt.legend()
plt.show()
```

```python
# Print MSE for training and testing
print(f'MSE for Training Data: {mse_train}')
print(f'MSE for Testing Data: {mse_test}')
```
